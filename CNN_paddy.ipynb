```python

import os
from os import walk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import random
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import f1_score, classification_report
from sklearn.utils.class_weight import compute_class_weight
import torch
from torch import nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
from torchvision import models, transforms
import timm
from tqdm import tqdm
from torch.cuda.amp import autocast, GradScaler

# Set seeds for reproducibility
def seed_everything(seed):
    random.seed(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

# Hyperparameters with improved values
HP = {
    'epochs': 25,
    'batch_size': 32,
    'learning_rate': 2e-4,  # Lower LR for finetuning pretrained model
    'weight_decay': 1e-4,   # L2 regularization
    'val_size': 0.15,       # Validation set size
    'test_size': 0.05,      # Test set size
    'seed': 42,
    'img_size': 224,
    'num_workers': 4,
    'use_metadata': True,   # Whether to use variety and age features
    'model_name': 'efficientnet_b0',  # Start with B0 for faster training
    'use_amp': True,        # Mixed precision training
    'early_stopping': 5,    # Stop if no improvement for 5 epochs
    'use_tta': True,        # Test-time augmentation
    'save_dir': '/kaggle/working/'
}

seed_everything(HP['seed'])
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f'Using {device} device')

# Paths
dataset_dir = '/kaggle/input/paddy-disease-classification/train_images/'
submission_dir = '/kaggle/input/paddy-disease-classification/test_images/'
dataset_file = '/kaggle/input/paddy-disease-classification/train.csv'
submission_sample = '/kaggle/input/paddy-disease-classification/sample_submission.csv'
submission_output = '/kaggle/working/submission.csv'

# Load and preprocess data
df = pd.read_csv(dataset_file)
df = shuffle(df, random_state=HP['seed'])

print(f'Total samples: {len(df)}')
print(f'Classes: {df.label.nunique()}')

# Better data visualization
print("\nClass distribution:")
class_counts = df['label'].value_counts()
print(class_counts)

print("\nVariety distribution:")
variety_counts = df['variety'].value_counts()
print(variety_counts)

# Create label mappings
labels = sorted(df['label'].unique())
label_to_idx = {label: idx for idx, label in enumerate(labels)}
idx_to_label = {idx: label for idx, label in enumerate(labels)}
print("Label to index mapping:", label_to_idx)

# Enhanced data augmentation
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(HP['img_size'], scale=(0.7, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(30),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.RandomPerspective(distortion_scale=0.2, p=0.5),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.2, scale=(0.02, 0.2))
])

val_transform = transforms.Compose([
    transforms.Resize(HP['img_size'] + 32),  # Resize larger then center crop
    transforms.CenterCrop(HP['img_size']),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Enhanced dataset with metadata
class PaddyDatasetWithMetadata(Dataset):
    def __init__(self, dataset_dir, df, label_to_idx, transforms, use_metadata=True, is_test=False):
        self.df = df.copy()
        self.label_to_idx = label_to_idx
        self.transforms = transforms
        self.use_metadata = use_metadata
        self.is_test = is_test

        # Preprocess paths
        if not is_test:
            self.df['path'] = dataset_dir + self.df.label + '/' + self.df.image_id
        else:
            self.df['path'] = dataset_dir + self.df.image_id

        # Encode variety as categorical
        if use_metadata:
            self.varieties = sorted(df['variety'].unique())
            self.variety_to_idx = {var: i for i, var in enumerate(self.varieties)}
            self.df['variety_idx'] = self.df['variety'].map(self.variety_to_idx)

            # Normalize age to 0-1 range (assume max age is 150 days)
            self.max_age = 150
            self.df['age_norm'] = self.df['age'].fillna(self.df['age'].median()) / self.max_age

        # Convert to list for faster access
        self.df = self.df.values.tolist()

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df[idx]
        # Extract appropriate columns based on is_test and use_metadata
        if self.is_test:
            image_id, _, variety, age, path = row[:5]
            label_idx = 0  # Placeholder
        else:
            image_id, label, variety, age, path = row[:5]
            label_idx = self.label_to_idx[label]

        # Load and transform image
        image = Image.open(path)
        image = self.transforms(image)

        # If using metadata
        if self.use_metadata:
            # One-hot encode variety
            variety_idx = self.variety_to_idx.get(variety, 0)
            variety_onehot = torch.zeros(len(self.varieties))
            variety_onehot[variety_idx] = 1

            # Handle age (normalize to 0-1)
            if pd.notna(age):
                age_norm = float(age) / self.max_age
            else:
                age_norm = 0.5  # Default value if missing

            return image, label_idx, variety_onehot, torch.tensor([age_norm], dtype=torch.float)
        else:
            return image, label_idx

# Create train/val/test splits with stratification
train_df, temp_df = train_test_split(df, test_size=HP['val_size'] + HP['test_size'],
                                     random_state=HP['seed'], stratify=df['label'])
val_df, test_df = train_test_split(temp_df, test_size=HP['test_size']/(HP['val_size'] + HP['test_size']),
                                   random_state=HP['seed'], stratify=temp_df['label'])

print(f"Train: {len(train_df)}, Validation: {len(val_df)}, Test: {len(test_df)}")

# Create datasets
train_dataset = PaddyDatasetWithMetadata(dataset_dir, train_df, label_to_idx, train_transform, HP['use_metadata'])
val_dataset = PaddyDatasetWithMetadata(dataset_dir, val_df, label_to_idx, val_transform, HP['use_metadata'])
test_dataset = PaddyDatasetWithMetadata(dataset_dir, test_df, label_to_idx, val_transform, HP['use_metadata'])

# Create dataloaders
train_dataloader = DataLoader(train_dataset, batch_size=HP['batch_size'], shuffle=True,
                             pin_memory=True, num_workers=HP['num_workers'])
val_dataloader = DataLoader(val_dataset, batch_size=HP['batch_size'], shuffle=False,
                           pin_memory=True, num_workers=HP['num_workers'])
test_dataloader = DataLoader(test_dataset, batch_size=HP['batch_size'], shuffle=False,
                            pin_memory=True, num_workers=HP['num_workers'])

# Compute class weights to handle imbalance
class_counts = train_df['label'].value_counts().sort_index()
class_weights = compute_class_weight('balanced', classes=np.unique(train_df['label'].map(label_to_idx)),
                                     y=train_df['label'].map(label_to_idx))
class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

# Create model with metadata integration
class PaddyModel(nn.Module):
    def __init__(self, model_name, num_classes, num_varieties=0, use_metadata=True):
        super(PaddyModel, self).__init__()
        self.use_metadata = use_metadata

        # Image feature extractor (backbone)
        if model_name.startswith('efficientnet'):
            self.backbone = timm.create_model(model_name, pretrained=True)
            self.feature_dim = self.backbone.classifier.in_features
            self.backbone.classifier = nn.Identity()
        elif model_name.startswith('resnet'):
            if model_name == 'resnet34':
                self.backbone = models.resnet34(pretrained=True)
            else:
                self.backbone = models.resnet50(pretrained=True)
            self.feature_dim = self.backbone.fc.in_features
            self.backbone.fc = nn.Identity()
        else:
            raise ValueError(f"Unsupported model name: {model_name}")

        # Metadata processing
        if use_metadata:
            self.metadata_dim = num_varieties + 1  # varieties + age
            self.metadata_encoder = nn.Sequential(
                nn.Linear(self.metadata_dim, 64),
                nn.BatchNorm1d(64),
                nn.ReLU(),
                nn.Dropout(0.3),
                nn.Linear(64, 32),
                nn.ReLU()
            )
            classifier_input = self.feature_dim + 32
        else:
            classifier_input = self.feature_dim

        # Classifier head
        self.classifier = nn.Sequential(
            nn.Dropout(0.4),
            nn.Linear(classifier_input, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, num_classes)
        )

    def forward(self, x, variety=None, age=None):
        # Extract features from image
        img_features = self.backbone(x)

        # Process metadata if enabled
        if self.use_metadata and variety is not None and age is not None:
            metadata = torch.cat([variety, age], dim=1)
            metadata_features = self.metadata_encoder(metadata)

            # Combine image and metadata features
            features = torch.cat([img_features, metadata_features], dim=1)
        else:
            features = img_features

        # Classification
        output = self.classifier(features)
        return output

# Create model
if HP['use_metadata']:
    num_varieties = len(train_df['variety'].unique())
    model = PaddyModel(HP['model_name'], len(label_to_idx), num_varieties, HP['use_metadata'])
else:
    model = PaddyModel(HP['model_name'], len(label_to_idx), use_metadata=False)

model = model.to(device)
print(f"Created model: {HP['model_name']} with metadata: {HP['use_metadata']}")

# Improved optimizer and loss
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = torch.optim.AdamW(model.parameters(), lr=HP['learning_rate'], weight_decay=HP['weight_decay'])

# Learning rate scheduler - OneCycle policy
scheduler = torch.optim.lr_scheduler.OneCycleLR(
    optimizer,
    max_lr=HP['learning_rate'],
    steps_per_epoch=len(train_dataloader),
    epochs=HP['epochs'],
    pct_start=0.2,  # 20% warmup
    div_factor=25,   # initial_lr = max_lr/div_factor
    final_div_factor=10000  # min_lr = initial_lr/final_div_factor
)

# Training function with mixed precision and validation
def train_val_epoch(model, train_loader, val_loader, criterion, optimizer, scheduler, epoch, use_amp=True):
    # Initialize metrics
    train_loss = 0
    train_correct = 0
    train_total = 0
    val_loss = 0
    val_correct = 0
    val_total = 0
    val_preds = []
    val_targets = []

    # Training phase
    model.train()
    scaler = GradScaler(enabled=use_amp)

    with tqdm(train_loader, unit='batch', desc=f"Epoch {epoch+1}/{HP['epochs']} [Train]") as pbar:
        for batch in pbar:
            if HP['use_metadata']:
                images, labels, variety, age = [item.to(device) for item in batch]
            else:
                images, labels = [item.to(device) for item in batch]
                variety, age = None, None

            optimizer.zero_grad(set_to_none=True)

            # Forward pass with mixed precision
            with autocast(enabled=use_amp):
                outputs = model(images, variety, age) if HP['use_metadata'] else model(images)
                loss = criterion(outputs, labels)

            # Backward and optimize with gradient scaling
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            # Update learning rate
            scheduler.step()

            # Update metrics
            train_loss += loss.item()
            _, predicted = outputs.max(1)
            train_total += labels.size(0)
            train_correct += predicted.eq(labels).sum().item()

            # Update progress bar
            train_acc = 100. * train_correct / train_total
            pbar.set_postfix(loss=f"{loss.item():.4f}", acc=f"{train_acc:.2f}%")

    # Validation phase
    model.eval()

    with torch.no_grad():
        with tqdm(val_loader, unit='batch', desc=f"Epoch {epoch+1}/{HP['epochs']} [Val]") as pbar:
            for batch in pbar:
                if HP['use_metadata']:
                    images, labels, variety, age = [item.to(device) for item in batch]
                else:
                    images, labels = [item.to(device) for item in batch]
                    variety, age = None, None

                # Forward pass
                outputs = model(images, variety, age) if HP['use_metadata'] else model(images)
                loss = criterion(outputs, labels)

                # Update metrics
                val_loss += loss.item()
                _, predicted = outputs.max(1)
                val_total += labels.size(0)
                val_correct += predicted.eq(labels).sum().item()

                # Collect predictions and targets for F1 score
                val_preds.extend(predicted.cpu().numpy())
                val_targets.extend(labels.cpu().numpy())

                # Update progress bar
                val_acc = 100. * val_correct / val_total
                pbar.set_postfix(loss=f"{loss.item():.4f}", acc=f"{val_acc:.2f}%")

    # Calculate metrics
    train_loss = train_loss / len(train_loader)
    train_acc = 100. * train_correct / train_total
    val_loss = val_loss / len(val_loader)
    val_acc = 100. * val_correct / val_total
    val_f1 = f1_score(val_targets, val_preds, average='macro')

    return {
        'train_loss': train_loss,
        'train_acc': train_acc,
        'val_loss': val_loss,
        'val_acc': val_acc,
        'val_f1': val_f1
    }

# Test Time Augmentation (TTA)
def tta_predict(model, image_path, num_augments=10):
    # Create multiple augmented versions of the image
    tta_transforms = [
        transforms.Compose([
            transforms.Resize(HP['img_size'] + 32),
            transforms.CenterCrop(HP['img_size']),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ]),
        transforms.Compose([
            transforms.Resize(HP['img_size'] + 32),
            transforms.CenterCrop(HP['img_size']),
            transforms.RandomHorizontalFlip(p=1.0),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ]),
        transforms.Compose([
            transforms.Resize(HP['img_size'] + 32),
            transforms.CenterCrop(HP['img_size']),
            transforms.RandomVerticalFlip(p=1.0),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ]),
        transforms.Compose([
            transforms.RandomResizedCrop(HP['img_size'], scale=(0.9, 1.0)),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ]),
        transforms.Compose([
            transforms.Resize(HP['img_size'] + 32),
            transforms.CenterCrop(HP['img_size']),
            transforms.ColorJitter(brightness=0.1),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ])
    ]

    # Load image
    image = Image.open(image_path).convert('RGB')

    # Get predictions for each augmentation
    model.eval()
    all_preds = []

    with torch.no_grad():
        # If using metadata, create default values (mode for varieties, median for age)
        if HP['use_metadata']:
            variety = torch.zeros(len(train_df['variety'].unique())).to(device)
            variety[0] = 1  # Set to most common variety
            age = torch.tensor([[0.5]]).to(device)  # Set to median age

        # Apply each transform multiple times
        for transform in tta_transforms:
            img_tensor = transform(image).unsqueeze(0).to(device)

            # Forward pass
            if HP['use_metadata']:
                outputs = model(img_tensor, variety, age)
            else:
                outputs = model(img_tensor)

            # Get softmax probabilities
            probs = F.softmax(outputs, dim=1)
            all_preds.append(probs.cpu())

    # Average predictions from all augmentations
    avg_preds = torch.mean(torch.cat(all_preds, dim=0), dim=0)
    pred_idx = torch.argmax(avg_preds).item()

    return pred_idx

# Training loop with early stopping
best_val_acc = 0
best_model_path = os.path.join(HP['save_dir'], f"best_{HP['model_name']}.pth")
early_stop_counter = 0
history = []

print("Starting training...")
for epoch in range(HP['epochs']):
    # Train and validate for one epoch
    metrics = train_val_epoch(model, train_dataloader, val_dataloader, criterion, optimizer, scheduler, epoch, HP['use_amp'])
    history.append(metrics)

    # Print metrics
    print(f"Epoch {epoch+1}/{HP['epochs']}")
    print(f"Train Loss: {metrics['train_loss']:.4f}, Train Acc: {metrics['train_acc']:.2f}%")
    print(f"Val Loss: {metrics['val_loss']:.4f}, Val Acc: {metrics['val_acc']:.2f}%, F1: {metrics['val_f1']:.4f}")

    # Save best model
    if metrics['val_acc'] > best_val_acc:
        best_val_acc = metrics['val_acc']
        torch.save(model.state_dict(), best_model_path)
        print(f"Saved new best model with Val Acc: {best_val_acc:.2f}%")
        early_stop_counter = 0
    else:
        early_stop_counter += 1
        print(f"Early stopping counter: {early_stop_counter}/{HP['early_stopping']}")

        if early_stop_counter >= HP['early_stopping']:
            print("Early stopping triggered!")
            break

# Plot training history
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot([x['train_loss'] for x in history], label='Train Loss')
plt.plot([x['val_loss'] for x in history], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot([x['train_acc'] for x in history], label='Train Acc')
plt.plot([x['val_acc'] for x in history], label='Val Acc')
plt.plot([x['val_f1'] for x in history], label='Val F1')
plt.xlabel('Epoch')
plt.ylabel('Accuracy / F1')
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(HP['save_dir'], 'training_history.png'))
plt.show()

# Load best model for inference
print("Loading best model for inference...")
model.load_state_dict(torch.load(best_model_path))
model.eval()

# Prepare submission data
print("Generating predictions for test data...")
sample_submission = pd.read_csv(submission_sample)
image_ids = []
labels = []

# Use TTA for predictions if enabled
for (dirpath, dirname, filenames) in walk(submission_dir):
    for filename in tqdm(filenames, desc="Predicting"):
        if not filename.endswith('.jpg'):
            continue

        image_path = os.path.join(dirpath, filename)
        image_id = filename.replace('.jpg', '')

        if HP['use_tta']:
            # Use test-time augmentation
            pred_idx = tta_predict(model, image_path)
        else:
            # Standard prediction
            image = Image.open(image_path).convert('RGB')
            image = val_transform(image).unsqueeze(0).to(device)

            with torch.no_grad():
                if HP['use_metadata']:
                    variety = torch.zeros(len(train_df['variety'].unique())).to(device)
                    variety[0] = 1  # Most common variety
                    age = torch.tensor([[0.5]]).to(device)  # Median age
                    outputs = model(image, variety, age)
                else:
                    outputs = model(image)

                pred_idx = outputs.argmax().item()

        # Get label from index
        pred_label = idx_to_label[pred_idx]

        image_ids.append(image_id)
        labels.append(pred_label)

# Create and save submission file
submission_df = pd.DataFrame({
    'image_id': image_ids,
    'label': labels
})

submission_df.to_csv(submission_output, index=False)
print(f"Submission saved to {submission_output}")
image_ids, labels = [], []
for (dirpath, dirname, filenames) in walk(submission_dir):
    for filename in filenames:
        image = Image.open(dirpath+filename)
        image = test_transform(image)
        image = image.unsqueeze(0).to(device)
        image_ids.append(filename)
        labels.append(idx_to_label[model(image).argmax().item()])
```
